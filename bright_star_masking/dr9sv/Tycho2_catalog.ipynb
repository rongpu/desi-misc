```python
from __future__ import division, print_function
import sys, os, glob, time, warnings, gc
import matplotlib.pyplot as plt
import numpy as np
from astropy.table import Table, vstack, hstack
import fitsio
```

```python
params = {'legend.fontsize': 'large',
         'axes.labelsize': 'large',
         'axes.titlesize':'large',
         'xtick.labelsize':'large',
         'ytick.labelsize':'large',
         'figure.facecolor':'w'} 
plt.rcParams.update(params)
```

```python
# Load trimmed Tycho-2 bright star catalog
tycho2 = Table.read('/global/project/projectdirs/desi/users/rongpu/useful/Tycho-2.fits')
print(len(tycho2))

tycho2.rename_column('RAmdeg', 'RA')
tycho2.rename_column('DEmdeg', 'DEC')
```

```python
plt.hist(tycho2['VTmag'][np.isfinite(tycho2['VTmag'])], 100)
plt.show()

plt.hist(tycho2['VTmag'][np.isfinite(tycho2['VTmag'])], 100, log=True)
plt.show()
```

```python
mask = tycho2['VTmag']<12
tycho2 = tycho2[mask]
print(len(tycho2))
```

```python
plt.figure(figsize=(18, 12))
plt.plot(tycho2['RA'][::5], tycho2['DEC'][::5], '.', markersize=0.3, alpha=0.5)
plt.axis([0, 360, -23, 80])
plt.grid(alpha=0.5)
plt.show()
```
