```python
from __future__ import division, print_function
import sys, os, glob, time, warnings, gc
import numpy as np
# import matplotlib
# matplotlib.use("Agg")
import matplotlib.pyplot as plt
from astropy.table import Table, vstack, hstack, join
import fitsio
# from astropy.io import fits
```

```python
params = {'legend.fontsize': 'x-large',
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large',
         'figure.facecolor':'w'} 
plt.rcParams.update(params)
```

```python
fns = [
'rrtemplate-star-B.fits',
'rrtemplate-star-A.fits',
'rrtemplate-star-F.fits',
'rrtemplate-star-G.fits',
'rrtemplate-star-K.fits',
'rrtemplate-star-M.fits',
]

plt.figure(figsize=(30, 15))
for fn in fns:
    d = fitsio.read('/Users/rongpu/moregit/redrock-templates/'+fn)
    if -np.min(d[0])>np.max(d[0]):
        d[0] = -d[0]
    wave = np.linspace(3000.0, 3000.0+0.1*(len(d[0])-1), len(d[0]))
    plt.plot(wave, d[0]/d[0].max(), lw=1, alpha=0.5, label=fn.replace('.fits', ''))
plt.legend()
plt.grid(alpha=0.5)
plt.grid(alpha=0.5)
plt.axis([wave.min(), wave.max(), -0.01, 1])
plt.xlabel('wavelength ($\AA$)')
plt.show()
```

```python
fn = 'rrtemplate-galaxy.fits'

idx_list = [np.arange(10)]
for idx in idx_list:
    plt.figure(figsize=(30, 15))
    d = fitsio.read('/Users/rongpu/moregit/redrock-templates/'+fn)
    for index in idx:
        if -np.min(d[index])>np.max(d[index]):
            d[index] = -d[index]
        wave = np.linspace(3000.0, 3000.0+0.1*(len(d[index])-1), len(d[index]))
        plt.plot(wave, d[index]/np.maximum(d[index].max(), np.abs(d[index].min())), lw=1, alpha=1, label='galaxy-pca-{}'.format(index))
    plt.legend()
    plt.grid(alpha=0.5)
    plt.grid(alpha=0.5)
    plt.axis([wave.min(), wave.max(), -1.05, 1.05])
    plt.xlabel('wavelength ($\AA$)')
    plt.show()
```
