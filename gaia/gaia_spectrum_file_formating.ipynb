```python
from __future__ import division, print_function
import sys, os, glob, time, warnings, gc
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table, vstack, hstack, join, Column
import fitsio

from astroquery.gaia import Gaia
```

```python
ttt = Table.read('/Users/rongpu/Downloads/XpContinuousMeanSpectrum_763561-763702.csv.gz', format='ascii.ecsv')
```

```python
tmp = ttt.copy()
```

```python
tmp = ttt.copy()

# convert int8 to int16 because int8 is not supported by FITS
for col in tmp.colnames:
    if tmp[col].dtype==np.int8:
        tmp[col] = np.array(tmp[col]).astype(np.int16)

# convert "object" type to numpy array
for col in ['bp_coefficients', 'bp_coefficient_errors', 'bp_coefficient_correlations', 'rp_coefficients', 'rp_coefficient_errors', 'rp_coefficient_correlations']:
    tmp[col] = np.array(list(tmp[col]))

tmp.write('/Users/rongpu/Downloads/XpContinuousMeanSpectrum_763561-763702.fits', overwrite=True)
```

```python
cat = Table(fitsio.read('/Users/rongpu/Downloads/XpContinuousMeanSpectrum_763561-763702.fits'))
print(len(cat))

# Workaround to make the multidimensional arrays pandas-compatible
cat = vstack([cat, cat[:1].copy()])
for col in ['bp_coefficients', 'bp_coefficient_errors', 'bp_coefficient_correlations', 'rp_coefficients', 'rp_coefficient_errors', 'rp_coefficient_correlations']:
    tmp = list(np.array(cat[col]))
    tmp.pop()
    tmp += [np.array([0])]
    cat[col] = tmp
cat = cat[:-1]
print(len(cat))
```

```python
pp = cat[:100].copy()
pp = pp.to_pandas()
```
